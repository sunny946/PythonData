```python
import pandas as pd
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
df = pd.read_csv('WorldIndex.csv')
df.head()
```

```python
df.info()
```

```python
df.Continent.unique()
```

```python
df.Continent.value_counts()
```

```python
df.plot(kind = 'scatter', x = 'GDP_per_capita', y = 'Life_expectancy')
```

```python
Africa = df[df.Continent == 'Africa']
Asia = df[df.Continent == 'Asia']
Europe = df[df.Continent == 'Europe']
North_America = df[df.Continent == 'North America']
Oceania = df[df.Continent == 'Oceania']
South_America = df[df.Continent == 'South America']

ax = Africa.plot(kind='scatter', x="GDP_per_capita", y="Life_expectancy", color='Red', label='Africa', figsize=(10,6))
Asia.plot(kind='scatter', x="GDP_per_capita", y="Life_expectancy", color='Green', ax=ax, label='Asia')
Europe.plot(kind='scatter', x="GDP_per_capita", y="Life_expectancy", color='orange', ax=ax, label='Europe')
North_America.plot(kind='scatter', x="GDP_per_capita", y="Life_expectancy", color='pink', ax=ax, 
                   label='North_America')
Oceania.plot(kind='scatter', x="GDP_per_capita", y="Life_expectancy", color='blue', ax=ax, label='Oceania')
South_America.plot(kind='scatter', x="GDP_per_capita", y="Life_expectancy", color='black', ax=ax, 
                   label='South_America')
```

```python
df[['Life_expectancy', 'Continent']].boxplot(grid=False, by='Continent', figsize=(10, 6))
```

```python
df[['GDP_per_capita', 'Continent']].boxplot(grid=False, by='Continent', figsize=(10, 6))
```

从第一张散点图大致趋势来看，在全球范围内，人均GDP越高，人们的期望寿命越大。

从第二张散点图来看，Europe的许多国家的人均GDP和期望寿命都要远远领先其他洲的其他国家；另外，当人均GDP超过15000之后，各个洲各自区域的期望寿命相差较小，大概在5年之内。Africa人均GDP和期望寿命正相关性不是很明显，应该还有其他影响人们期望寿命的重要相关因素。

两张箱图，从洲与洲的数据的比较来看，两张图的走势形态相似，比如两张图中，Africa的情况都是最糟的，Europe的情况都是最好的，这样也可以说明在全球范围内，人均GDP越高，人们的期望寿命越大。

从第一张箱图来看，Africa和Asia的期望寿命最大值和最小值相差很大，大概差了二十几年，其中原因值得探寻。

从第二张箱图来看，Europe不同国家之间贫富差距较大。
